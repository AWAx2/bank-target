```python
import pandas as pd
```

```python
pred_best = pd.read_csv('submit/submit_30_1438.csv', names=('id', 'best'))
pred_second = pd.read_csv('submit/submit_30_1444.csv', names=('id', 'second'))
```

```python
pred = pd.merge(pred_best, pred_second, on='id')
pred.head(-10)
```

```python
pred['mean'] = pred[['best', 'second']].mean(axis=1)
pred['diff'] = pred.best - pred.second
```

```python
diff_min = pred['diff'].min()
diff_max = pred['diff'].max()
diff_mean = pred['diff'].mean()

print('min: {}, max: {}, mean:{}'.format(diff_min, diff_max, diff_mean))
```
